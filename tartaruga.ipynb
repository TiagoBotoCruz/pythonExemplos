```python
import turtle as tt
import random as rd
```

```python
tela = tt.Screen()
tt.title("Tartarugas Ninjas")

miquelangelo = tt.Turtle()
donatelo = tt.Turtle()
rafael = tt.Turtle()
leo = tt.Turtle()
narrador = tt.Turtle()

narrador.write('Coloque cada tartaruga em seu respectivo círculo')

leo.shape("turtle")
miquelangelo.shape("turtle")
donatelo.shape("turtle")
rafael.shape("turtle")

leo.color("blue")
miquelangelo.color("yellow")
donatelo.color("purple")
rafael.color("red")

leo.penup()
miquelangelo.penup()
donatelo.penup()
rafael.penup()

leo.goto(50,300)
miquelangelo.goto(200,300)
donatelo.goto(-50,-300)
rafael.goto(-280,-300)

leo.pendown()
miquelangelo.pendown()
donatelo.pendown()
rafael.pendown()

leo.pen(pencolor="blue", fillcolor="yellow", pensize=5, speed=1)
leo.begin_fill()
leo.circle(50)
leo.end_fill()

miquelangelo.pen(pencolor="yellow", fillcolor="purple", pensize=5, speed=1)
miquelangelo.begin_fill()
miquelangelo.circle(50)
miquelangelo.end_fill()

donatelo.pen(pencolor="purple", fillcolor="red", pensize=5, speed=1)
donatelo.begin_fill()
donatelo.circle(50)
donatelo.end_fill()

rafael.pen(pencolor="red", fillcolor="blue", pensize=5, speed=1)
rafael.begin_fill()
rafael.circle(50)
rafael.end_fill()

leo.penup()
miquelangelo.penup()
donatelo.penup()
rafael.penup()

leo.goto(-50,0)
miquelangelo.goto(-15,0)
donatelo.goto(50,0)
rafael.goto(15,0)

def esquerdaLeo():
    leo.setx(leo.xcor() - 5)
    
def direitaLeo():
    leo.setx(leo.xcor() + 5)

def cimaLeo():
    leo.sety(leo.ycor() + 5)

def baixoLeo():
    leo.sety(leo.ycor() - 5)

def esquerdaMiq():
    miquelangelo.setx(miquelangelo.xcor() - 5)
    
def direitaMiq():
    miquelangelo.setx(miquelangelo.xcor() + 5)

def cimaMiq():
    miquelangelo.sety(miquelangelo.ycor() + 5)

def baixoMiq():
    miquelangelo.sety(miquelangelo.ycor() - 5)


def esquerdaDon():
    donatelo.setx(donatelo.xcor() - 5)
    
def direitaDon():
    donatelo.setx(donatelo.xcor() + 5)

def cimaDon():
    donatelo.sety(donatelo.ycor() + 5)

def baixoDon():
    donatelo.sety(donatelo.ycor() - 5)
    
def esquerdaRaf():
    rafael.setx(rafael.xcor() - 5)
    
def direitaRaf():
    rafael.setx(rafael.xcor() + 5)

def cimaRaf():
    rafael.sety(rafael.ycor() + 5)

def baixoRaf():
    rafael.sety(rafael.ycor() - 5)
    
tela.onkeypress(esquerdaLeo,'Left')
tela.onkeypress(direitaLeo,'Right')
tela.onkeypress(cimaLeo,'Up')
tela.onkeypress(baixoLeo,'Down')

tela.onkeypress(esquerdaMiq,'a')
tela.onkeypress(direitaMiq, 'd')
tela.onkeypress(cimaMiq, 'w')
tela.onkeypress(baixoMiq, 's')

tela.onkeypress(esquerdaDon,'f')
tela.onkeypress(direitaDon, 'h')
tela.onkeypress(cimaDon, 't')
tela.onkeypress(baixoDon, 'g')

tela.onkeypress(esquerdaRaf,'j')
tela.onkeypress(direitaRaf, 'l')
tela.onkeypress(cimaRaf, 'i')
tela.onkeypress(baixoRaf, 'k')

tela.listen()

tt.done()
```

```python
tela2 = tt.Screen()
tela2.title('Desenho')
tela2.setup(width=720, height=720)

obj1 = tt.Turtle()
obj1.pensize(8)
obj1.speed(5)

def clique(x,y):
    obj1.goto(x,y)
    
tela2.onscreenclick(clique)
tela2.listen()

tt.done()
```

```python
tela3 = tt.Screen()
tela3.title('Treino de Mouse')
tela3.setup(width=720, height=720)

obj2 = tt.Turtle()
obj2.pensize(8)
obj2.speed(5)
obj2.penup()
obj2.write('clique nos cinco círculos')

obj3 = tt.Turtle()
obj3.penup()

for i in range(5):
    x = rd.randint(-300,300)
    y = rd.randint(-300,300)
    obj3.speed(1)
    obj3.goto(x,y)
    obj3.pendown()
    obj3.pen(pencolor="blue", fillcolor="blue", pensize=1, speed=1)
    obj3.begin_fill()
    obj3.circle(25)
    obj3.end_fill()
    obj3.penup()

def clique(x,y):
    obj2.goto(x,y)
    
tela3.onscreenclick(clique)
tela3.listen()

tt.done()
```
